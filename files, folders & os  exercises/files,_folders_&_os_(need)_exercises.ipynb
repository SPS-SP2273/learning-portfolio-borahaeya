TASK 1

````python
month_mapping = {
    'jan': '01', 'feb': '02', 'mar': '03',
    'apr': '04', 'may': '05', 'jun': '06',
    'jul': '07', 'aug': '08', 'sep': '09',
    'oct': '10', 'nov': '11', 'dec': '12'
}

file_name = 'oct-08_singapore.txt'

# split into parts based on underscores
parts = file_name.split('_')
#oct-08 + singapore.txt

month_abbreviation = parts[0][:3] #extract first element (oct-08) of parts and its first 3 characters (oct)
if month_abbreviation in month_mapping: #abbreviation of oct in month_mapping print its corresponding number
    month = month_mapping[month_abbreviation]
else:
    month = month_abbreviation

# new file name
new_file_name = f"{month}-{parts[0][4:]}-{parts[1]}"
# parts[0] oct-08 
# parts[0][4:] from 4th character onwards in oct-08 -> -08
# parts[1] singapore.txt
print(new_file_name)
````

TASK 2

````python
def rename_my_file(old_file_name): #old_file_name new variable so can apply to all files names
    parts = old_file_name.split('_')

    month_abbreviation = parts[0][:3] 
    if month_abbreviation in month_mapping: 
        month = month_mapping[month_abbreviation]
    else:
        month = month_abbreviation
    
    new_file_name = f"{month}-{parts[0][4:]}-{parts[1]}"
    return new_file_name
````

````python
rename_my_file('jul-10_cairo.txt')
````

````python
cities= ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']
for world in cities:
    globe= glob.glob(f'**/*{world.lower()}*.txt', recursive=True)
    for earth in globe:
        map= os.path.join('data_cities', world, earth.split(os.path.sep)[-1]) #this step to list new address to move files later
        shutil.move(earth, map)
        print(f'Moved file to {map}')
````
