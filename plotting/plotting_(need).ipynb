# 1 A simple plot

```python
from matplotlib import pyplot as plt
#need to import when opened as new
```

## 1.1 Letâ€™s look at some code.

```python
x= [0,1,2,3,4,5,6]
y_1= [0,2,4,6,8,10,12]
y_2= [0,3,6,9,12,15,18]
#DATA FOR PLOTTING

plt.plot(x,y_1,label= 'Y values', color='red', linestyle='dashed', linewidth=2)
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
#grid lines
#alpha- transparency of grid lines, the higher value the higher colour intensity and opaque
plt.legend(loc='upper left') #legend location upper left
#order of keyword arguments e.g. color, linewidth doesnt matter
```

## 1.3 Yet another plot but with error bars

```python
x= [0,1,2,3,4,5,6]
y_1= [0,2,4,6,8,10,12] # from 0, multiples of 2
y_2= [0,3,6,9,12,15,18] # from 0, multiples of 3
y_3= [0,4,8,12,16,20,24] # from 0, multiples of 4
x_error,y_error=.1,0.75 #error bar sizes

plt.plot(x,y_1,label= 'Y1 values', color='red', linestyle='dashed')
plt.plot(x,y_2,label='Y2 values',color='blue')

#
plt.errorbar(x,y_3,xerr=x_error,yerr=y_error,label='Y3 with errors',color='green')
# xerr x error bar, y error bar

plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25) 
plt.legend(loc='upper left') 
```

# 2 Better with NumPy

```python
import numpy as np
#numpy arrays better than python lists

x= [0,1,2,3,4,5,6]
y_1= [0,2,4,6,8,10,12] 
y_2= [0,3,6,9,12,15,18]

np_x=np.array(x)
np_y_1=np.array(y_1)
np_y_2=np.array(y_2)

plt.plot(np_x,np_y_1,label= 'Y1 values', color='red', linestyle='dashed')
plt.plot(np_x,np_y_2,label='Y2 values',color='blue')
```

## 2.1 Adding mathematical functions

```python
x=np.array([0,1,2,3,4,5,6])
x2=x**2 
sin_x=np.sin(x)

plt.plot(x,x2,label='x^2',color='red',linestyle='dashed')
plt.plot(x,sin_x,label='sin(x)',color='blue')
plt.legend()
```

## We need another axis!

```python
#twinx() have another y axis that shares the same x axis

x=np.array([0,1,2,3,4,5,6])
x2=x**2 
sin_x=np.sin(x)

plt.plot(x,x2,label='x^2',color='red',linestyle='dashed')
plt.legend(loc='lower left')

plt.twinx()

plt.plot(x,sin_x,label='sin(x)',color='blue')
plt.legend(loc='lower right')
```

### Things to note

```python
#make plot look nicer using np.linspace

x=np.array([0,1,2,3,4,5,6])
x=np.linspace(0,6,100) #start from 0, end on 6, generate 100 items
x2=x**2 
sin_x=np.sin(x)

plt.plot(x,x2,label='x^2',color='red',linestyle='dashed')
plt.legend(loc=(.25,.5), frameon=False) #specify legend location using coordinates

plt.twinx()

plt.plot(x,sin_x,label='sin(x)',color='blue')
plt.legend(loc=(.25,.45), frameon=False)

#frameon= False -> legend will not be surrounded by a coloured border
```

# 4 A real example: Global Warming

## 4.1 Plotting data from files

```python
data=np.loadtxt('Complete_TAVG_daily.txt',skiprows=24) #skiprows skip first 24 rows

date=data[:,0] #extract numbers of first column of every row
anomaly=data[:,-1] #extract numbers of last column of every row
#works as numpy arrays

plt.plot(date,anomaly,alpha=.5) #alpha to adjust opacity of plot
plt.ylim([-8,8]) #range of y axis, if x axis then xlim

plt.hlines(0,date[0],date[-1],linestyle='dashed',color='grey')
#hlines() add horizontal line at zero
#plt.hlines(y value, starting x value, ending x value)
```

## 4.2 Styles

```python
plt.style.use('fivethirtyeight')
#using matplotlib standard style templates
#this line must be at the top

date=data[:,0] 
anomaly=data[:,-1] 

plt.plot(date,anomaly,alpha=.5) 
plt.ylim([-8,8]) 

plt.hlines(0,date[0],date[-1],linestyle='dashed',color='grey')

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly/n(Relative to average from Jan 1951 - Dec 1980)')
```
