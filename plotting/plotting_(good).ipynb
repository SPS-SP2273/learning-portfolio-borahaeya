# 2 Comparing the two ‘dialects’

````python
import numpy as np

#data
x= np.linspace(-np.pi, np.pi, num=100)
#np.linspace return evenly spaced numbers over a specified interval
#np.pi returns pi as a floating point number
cos_x= np.cos(x)
sin_x= np.sin(x)

#plot 1
plt.plot(x, cos_x, label='cos x')
plt.plot(x, sin_x, label='sin x')
plt.legend()

#plot 2
#plt.subplots can take 3 arguments to describe figure layout 
#e.g. (1,2,1) 1 row 2 columns 1st plot
fig, ax= plt.subplots(nrows=1, ncols=1)
ax.plot(x, cos_x, label= 'cos x')
ax.plot(x, sin_x, label= 'sin x')
ax.legend()
````

# 3 What is OO ax, really?

````python
fig, ax= plt.subplots(nrows=2, ncols=1)
#plt.subplots() returns <Matplotlib Figure Object> and <numpy list of graph window objects>
#graph windows is <Matplotlib Axes Object>
#nrows= 2 -> 2 rows
#ncols= 1 -> 1 column
#gives 2 axes to plot in both panes
#fig, ax= plt.subplots() means assign <Matplotlib Figure Object> variable name fig & <numpy list> variable name ax

ax[0].plot(x, cos_x, label= 'cos x')
ax[1].plot(x, sin_x, label= 'sin x')

ax[0].legend()
ax[1].legend()
#index [0] display as 1st plot, [1] display as 2nd plot
````

# 4 A complete OO example

````python
#nicer version of plot in 4.3

fig, ax= plt.subplots(nrows=2, ncols=1, 
                      figsize= (5,5),
                      sharex= True)
#2 axes- 2 rows 1 column
#figure size- figsize
#sharex- all plots share the same x axis

ax[0].plot(x, cos_x, label= r'$\cos(x)$')
ax[0].fill_between(x,0,cos_x,alpha=.25)
ax[1].plot(x, sin_x, label= r'$\sin(x)$')
ax[1].fill_between(x,0,sin_x,alpha=.25)
#ax[].plot plot to one axis
#fill_between() to fill the plots

for a in ax: #use for loop to apply to both plots
    a.legend()
    a.grid(alpha=.25)
    a.set_ylabel('$y$', rotation=0)

ax[1].set_xlabel('$x$')
#sharing x axis only need to label lowest plot index 1

fig.suptitle(r'$\sin(x)$ and $\cos(x)$')
#suptitle- super title
fig.tight_layout()
#adjust subplot to fit into figure area nicely compact
````

````python
#pyplot and OO (object oriented) interface have different function names
#e.g. x label
#pyplot -> xlabel()
#OO (uses methods which are functions belonging to an object) -> set_xlabel()
````

# 5 Multiple rows and columns

````python
#using this whole plot for the rest of the section
````

````python
fig, ax= plt.subplots(nrows=2, ncols=2,
                     figsize=(5,5),
                     sharex='col', sharey='col')
#sharex= 'col' each column shares the same x axis
#sharey= 'col' each column shares the same y axis

#access axes improve readability
#4 axes and its locations
top_left, top_right, bottom_left, bottom_right= ax.flatten()

#set_title -> set plot titles
top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

#hlines -> set horizontal line 
#xmin xmax set x axis range of drawn line
top_left.hlines(y=1,xmin=0,xmax=4)
top_right.hlines(y=2,xmin=0,xmax=5)
bottom_left.hlines(y=3,xmin=0,xmax=4)
bottom_right.hlines(y=4,xmin=0,xmax=5)

for a in ax.flatten():
    a.grid(alpha=.25)

plt.tight_layout()
````

## 5.1 Sharing axes

````python
#without sharing axes
#auto scaled both axes
fig, ax= plt.subplots(nrows=2,ncols=2,figsize=(5,5))

top_left= ax[0,0]
top_right= ax[0,1]
bottom_left= ax[1,0]
bottom_right= ax[1,1]

top_left.hlines(y=1,xmin=0,xmax=4)
top_right.hlines(y=2,xmin=0,xmax=5)
bottom_left.hlines(y=3,xmin=0,xmax=4)
bottom_right.hlines(y=4,xmin=0,xmax=5)
````

````python
#sharing axes
#True= makes all axes use same range
#col= same range for all columns
#row= same range for all rows

fig, ax= plt.subplots(nrows=2,ncols=2,figsize=(5,5),sharex=True, sharey='row')

top_left= ax[0,0]
top_right= ax[0,1]
bottom_left= ax[1,0]
bottom_right= ax[1,1]

top_left.hlines(y=1,xmin=0,xmax=4)
top_right.hlines(y=2,xmin=0,xmax=5)
bottom_left.hlines(y=3,xmin=0,xmax=4)
bottom_right.hlines(y=4,xmin=0,xmax=5)

#sharex= True all axes use same x axis range
#sharey= 'row' each row shares the same y axis range
````

````python
#sharex= 'col' more suited for data plotting here than = True
fig, ax= plt.subplots(nrows=2,ncols=2,figsize=(5,5),sharex= 'col', sharey='row')

top_left= ax[0,0]
top_right= ax[0,1]
bottom_left= ax[1,0]
bottom_right= ax[1,1]

top_left.hlines(y=1,xmin=0,xmax=4)
top_right.hlines(y=2,xmin=0,xmax=5)
bottom_left.hlines(y=3,xmin=0,xmax=4)
bottom_right.hlines(y=4,xmin=0,xmax=5)
````

## 6.1 Histograms

### Data

### A quick helper function

````python
def get_plot_data():
    data = {}
    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_all.csv'
    data['All'] = np.loadtxt(filename, skiprows=1)

    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_nus.csv'
    data['NUS'] = np.loadtxt(filename, skiprows=1)

    return data
````

### The histogram

````python
plt.style.use('bmh')
data = get_plot_data()

# bins specifies how many bins to split the data
plt.hist([data['All'], data['NUS']], bins=50, label=['All', 'NUS'])
plt.xlabel('Mean of Basic Montly Earning (S$)')
plt.ylabel('Number of Students')
plt.legend()
````

## 6.2 Scatter plots

````python
plt.style.use("seaborn-v0_8-darkgrid")

data= get_plot_data() #use same data for histogram

for label, numbers in data.items():
    x= range(len(numbers)) #generate a list appropriate to dataset
    y= numbers
    plt.scatter(x,y,label=label,alpha=.5)

plt.xlabel('Position in the list')
plt.ylabel('Mean of Basic Monthly Earning (S$)')
plt.legend()
````

## 6.3 Bar charts

````python
#hypothetical class
#pass 2 lists to bar()
#barh() for horizontal bars

studentnumber= {'Life Sciences':14,
               'Physics':12,
               'Chemistry':8,
               'Comp Biology':1}
majors= list(studentnumber.keys()) #keys -> majors
numbers= list(studentnumber.values()) #values -> number of students

plt.style.use('ggplot')
plt.bar(majors,numbers)
plt.xlabel('Majors')
plt.ylabel('Number of Students')
````

## 6.4 Pie charts

````python
studentnumber= {'Life Sciences':14,
               'Physics':12,
               'Chemistry':8,
               'Comp Biology':1}
majors= list(studentnumber.keys()) 
numbers= list(studentnumber.values())

plt.style.use('fivethirtyeight')
plt.pie(numbers,labels=majors,autopct='%1.1f%%',startangle=-90)
#autopct -> formating percentages
#startangle -> all slices rotated counterclockwise by 90°

plt.title('Percentage of each major')
````
